```python
import numpy as np
import pandas as pd
import cv2, glob, numpy
import math
%pylab inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
def scaleRadius(img,scale):
    x=img[int(img.shape[0]/2),:,:].sum(1)
    r=(x>x.mean()/10).sum()/2
    s=scale*1.0/r
    return cv2.resize(img,(0,0),fx=s,fy=s)
    return
```

```python

scale = 512

for f in (glob.glob("../Data/*.jpeg")):
# Display orignal image    
    img=mpimg.imread(f)
    imgplot = plt.imshow(img)
    plt.show()

#Scaling and contrast enhancement of the image 
    a=cv2.imread(f)
    a=scaleRadius(a,scale)
#   aa=cv2.addWeighted(a,4,cv2.GaussianBlur(a,(0,0),scale/30),-4,128)*b+128*(1-b)
    aa=cv2.addWeighted(a,4,cv2.GaussianBlur(a,(0,0),scale/30),-4,128)
    
# Constrast Enhanced image
    cv2.imwrite(str(scale)+f+".bmp",aa)
    plt.imshow(aa, interpolation='nearest')
    plt.show()
```
